### The `chr()` and `ord()` Functions

_`ord(c)`: Returns an integer representing the Unicode code point of the character `c`._

```python
ord('A')
```

_`chr(i)`: Returns a string of one character whose ASCII code is the integer `i`._

```python
chr(65)
```

```python
k = 3

chr(65 + k)
```

```python
ord(chr(65 + k))
```
